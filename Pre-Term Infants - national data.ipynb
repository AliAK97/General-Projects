### Handling Multiclass categorical variables

```python
for column in df.columns:
    if df[column].dtype == "category":
        print(f"- name of the column is: {column}")
        print(df[column].value_counts())
        print("\n")
```

```python
df["MomRes"] = df["MomRes"].astype("category")
df["MomRes"] = df["MomRes"].cat.set_categories(['شهر',
                                                'روستا',
                                                'عشایر'
                                                  ])

df.dropna(subset = "MomRes" , inplace = True)

mapping = {"شهر": "urban",
           "روستا" : "rural",
           "عشایر" : "rural"
          }

df["MomRes"] = df["MomRes"].map(mapping)

df["city/rural"] = pd.factorize(df["MomRes"])[0]
df = df.drop("MomRes", axis = 1)
df["city/rural"].value_counts()

# city == 1
# rural = 0
```

## Feature Scaling

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

scaled_columns = ColumnTransformer([('full', StandardScaler(), final_columns)])
X1 = scaled_columns.fit_transform(X)
```

```python
X1 = pd.DataFrame(X1)
```

```python
X1.head()
```

```python
X1.info()
```

## Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)
```

## Select and Train a Model on Training Set

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, \
                            precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

### Prediction of mortality in 24 hours

```python
models = {
    "Logistic Regression" : LogisticRegression(),
    "Random Forest" : RandomForestClassifier(),
    "GBM" : GradientBoostingClassifier(),
    "Naive Bayes" : GaussianNB()
}

for i in range(len(list(models))):
    model = list(models.values())[i]

    y_pred = cross_val_predict(model, X1, y1, cv=skf)

    model_accuracy = accuracy_score(y1, y_pred)
    model_precision = precision_score(y1, y_pred)
    model_recall = recall_score(y1, y_pred)
    model_f1 = f1_score(y1, y_pred, average = "weighted")
    model_rocauc = roc_auc_score(y1, y_pred)
    
    
    
    print(list(models.keys())[i])
    
    print("------------ \n")
    
    print("model performance:")
    print(f"- accuracy is: {model_accuracy: .4f}")
    print(f"- precision is: {model_precision: .4f}")
    print(f"- recall is: {model_recall: .4f}")
    print(f"- f1_score is: {model_f1: .4f}")
    print(f"- auc-roc score is: {model_rocauc: .4f}")
    
    
    print("-------------------------------------------------\n")
```
